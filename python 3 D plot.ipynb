```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax=plt.axes(projection='3d')
```

```python
fig = plt.figure()
ax=plt.axes(projection='3d')
zline = np.linspace(0, 15, 100)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'r-')
ax.set(xlabel='x values', ylabel='y values', zlabel='z values')
fig.set_size_inches(8,7)
plt.show()
```

```python
#Scatter plotting
fig = plt.figure()
ax=plt.axes(projection='3d')
zline = np.linspace(0, 15, 100)
xline = np.sin(zline)
yline = np.cos(zline)
ax.scatter3D(xline, yline, zline, c=zline, cmap='Greens')
ax.set(xlabel='x values', ylabel='y values', zlabel='z values')
fig.set_size_inches(8,7)
plt.show()
```

```python
#contour plot

def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
cp=ax.contour3D(X, Y, Z, 50, cmap='hot')
ax.set(xlabel='x values', ylabel='y values', zlabel='z values')
plt.colorbar(cp)
fig.set_size_inches(8,7)
plt.show()
```

```python
ax.view_init(60, 35)
fig
```

```python
#wireframe plotting
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X, Y, Z, color='black')
ax.set(xlabel='x values', ylabel='y values', zlabel='z values')
fig.set_size_inches(8,7)
plt.show()
```

```python
#surface plotting
fig=plt.figure()
ax = plt.axes(projection='3d')
sp=ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.set(xlabel='x values', ylabel='y values', zlabel='z values')
plt.colorbar(sp)
fig.set_size_inches(8,7)
plt.show()
```

```python
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))
z = f(x, y)
fig=plt.figure()
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z,
                cmap='viridis', edgecolor='none')
ax.set(xlabel='x values', ylabel='y values', zlabel='z values')
fig.set_size_inches(8,7)
plt.show()
```
